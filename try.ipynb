```python
from utils import normalize_audio
import librosa
import matplotlib.pyplot as plt
```

```python
audio1_path = './data_raw/TRAIN/118/47824/118-47824-0000.flac'
audio1,sr = librosa.load(audio1_path,sr=None)
```

```python
audio1_norm = normalize_audio(audio1)
plt.plot(audio1_norm)
```

```python
fig = plt.figure(figsize=(10, 7))

# 创建子图
ax1 = fig.add_axes([0, 0.5, 0.45, 0.45])  # 在图表的左上角创建一个子图
ax2 = fig.add_axes([0.5, 0, 0.45, 0.45])  # 在图表的右下方创建一个子图

# 左上角子图：曲线图
ax1.plot(audio1)

# 右下角子图：柱状图
ax2.plot(audio1_norm)
```
